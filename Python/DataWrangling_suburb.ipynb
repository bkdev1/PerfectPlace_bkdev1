```python
def DistanceToCity_label_func(x,y):
    if x<y[4]:
        return 'Very Near'
    elif x<y[5]:
        return 'Near'
    elif x<y[6]:
        return 'Far'
    else:
        return 'Very Far'
```

```python
def TravelTimeHospital_label_func(x,y):
    if x<y[4]:
        return 'Very Less'
    elif x<y[5]:
        return 'Less'
    elif x<y[6]:
        return 'Moderate'
    else:
        return 'High'
```

```python
def Generic_label_func(x,y):
    if x<y[4]:
        return 'Low'
    elif x<y[5]:
        return 'Moderate'
    elif x<y[6]:
        return 'High'
    else:
        return 'Very High'
```

```python
ratingdf = pd.DataFrame(index=range(331))
```

```python
ratingdf['community_name'] = subdf1.community_name
```

```python
ratingdf['distance_to_city'] = subdf1.apply(lambda row: DistanceToCity_label_func(row['distance_to_gpo'], subdf1.distance_to_gpo.describe()), axis=1)

ratingdf['residential'] = subdf1.apply(lambda row: Generic_label_func(row['residential_percent'], subdf1.residential_percent.describe()), axis=1)

ratingdf['health_services'] = health_col_df.apply(lambda row: Generic_label_func(row['health_col_list'], health_col_df.health_col_list.describe()), axis=1)

ratingdf['disability_care'] = subdf1.apply(lambda row: Generic_label_func(row['disability'], subdf1.disability.describe()), axis=1)

ratingdf['aged_care'] = agedcare_col_df.apply(lambda row: Generic_label_func(row['agedcare_col_list'], agedcare_col_df.agedcare_col_list.describe()), axis=1)

ratingdf['time_to_hospital'] = subdf1.apply(lambda row: TravelTimeHospital_label_func(row['travel_time_nearest_public_hospital'], subdf1.travel_time_nearest_public_hospital.describe()), axis=1)





```

```python
ratingdf['top_country_of_birth_1'] = subdf1.top_country_of_birth_1
ratingdf['top_country_of_birth_2'] = subdf1.top_country_of_birth_2
ratingdf['top_country_of_birth_3'] = subdf1.top_country_of_birth_3
ratingdf['top_country_of_birth_4'] = subdf1.top_country_of_birth_4
ratingdf['top_country_of_birth_5'] = subdf1.top_country_of_birth_5
```
