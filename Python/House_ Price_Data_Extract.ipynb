```python
from bs4 import BeautifulSoup
import requests
import webbrowser
import pandas as pd
```

```python
subdf = pd.read_csv('Suburb_Data_Full_v11.csv', sep=',')
```

```python
nmefunc = lambda x: x.lower().strip()
subdf['suburb'] = subdf.suburb.dropna().apply(nmefunc)
pricedf = subdf[['suburb_id','suburb']]
```

```python
pricedf['house_low_quartile']=0
pricedf['house_median_quartile']=0
pricedf['house_high_quartile']=0
pricedf['unit_low_quartile']=0
pricedf['unit_median_quartile']=0 
pricedf['unit_high_quartile']=0
```

```python
# tempdf1 = pd.DataFrame().T
# i=0
temp2 = 0
temp3 = 0
for index, row in pricedf.iterrows():
# def price(suburb):
 
#     i  = index+1
    print row['suburb']
    r  = requests.get("https://propertydata.realestateview.com.au/propertydata/median-prices/victoria/"+row['suburb']+"/")
    data = r.text
    soup = BeautifulSoup(data)
    i=0
    temparray = []
    print row['suburb']
    for row1 in soup.find_all('tr'):
        i = i+1
        if i==4:
            break
        else:
            temparray.append(row1.text)
    
    new_list = [temparray[i:i+1] for i in range(0, len(temparray), 1)]

    
    try:
        temp2 = [(w.replace(',', '')).replace('\nHouse\n$', '',1).replace('\n','').replace('$',' ') for w in  new_list[1]]
        temp2 = map(int,[word for line in temp2 for word in line.split()])
    except:
        temp2 = [0,0,0]
    try:   
        temp3 = [(w.replace(',', '')).replace('\nUnit\n $', '',1).replace('\n','').replace('$','') for w in  new_list[2]]
        temp3 = map(int,[word for line in temp3 for word in line.split()])
    except:
        temp3 = [0,0,0]
    temp = [index,row['suburb']] +temp2 + temp3
    
    print temp
    pricedf.loc[index] = temp
        
        
```

```python
def human_format(num):
    magnitude = 0
    while abs(num) >= 1000:
        magnitude += 1
        num /= 1000.0
    # add more suffixes if you need them
    return '%.2f%s' % (num, ['', 'K', 'M', 'G', 'T', 'P'][magnitude])

print('the answer is %s' % human_format(7436313))
```

```python
pricedf_temp['house_median_quartile_string'] = pricedf_temp['house_median_quartile'].apply(human_format)
pricedf_temp['unit_median_quartile_string'] = pricedf_temp['unit_median_quartile'].apply(human_format)
```
